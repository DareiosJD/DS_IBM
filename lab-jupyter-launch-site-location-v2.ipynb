First, let's try to add each site's location on a map using site's latitude and longitude coordinates

Now, you can take a look at what are the coordinates for each site.

Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Assuming your DataFrame is called launch_sites and has a 'class' column
launch_sites_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

Findings from Distance Analysis:
1. Are launch sites in close proximity to railways?

Yes.

    For example, Kennedy Space Center (KSC LC-39A) and Cape Canaveral (CCAFS) both have rail lines leading directly to the launch facilities.

    These are used for transporting large rocket components, fuel, and equipment.

ðŸ’¡ Why?

    Rockets and hardware are massive and require rail transport (too large or dangerous for roads alone).

    Rail provides cost-efficient logistics for heavy cargo.

2. Are launch sites in close proximity to highways?

Yes.

    All major US launch sites (Floridaâ€™s CCAFS, VAFB in California) have nearby highways or access roads.

    Example: VAFB SLC-4E connects to California State Route 1 (Pacific Coast Highway).

ðŸ’¡ Why?

    Highways allow staff, materials, and equipment to reach the site quickly.

    Emergency evacuation and logistics also benefit from road access.

3. Are launch sites in close proximity to coastline?

Definitely yes.

    All sites are within a few kilometers of the coast:

        CCAFS & KSC: right on the Atlantic.

        VAFB: right on the Pacific.

ðŸ’¡ Why?

    Safety: rockets launch eastward (to benefit from Earthâ€™s rotation), and failures or dropped stages fall into the ocean.

    No overflight of populated land, reducing risk.

4. Do launch sites keep a certain distance away from cities?

Yes, intentionally.

    KSC is about 50 km away from Orlando.

    VAFB is over 100 km from Los Angeles, and even the nearby town of Lompoc is small and distant.

ðŸ’¡ Why?

    Minimize risk to civilian populations in case of failure or explosion.

    Rocket launches produce shockwaves and toxic fumes â€” remote locations reduce harm.

    Noise and vibration can be significant.
